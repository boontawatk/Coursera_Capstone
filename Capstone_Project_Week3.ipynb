===================================================First Assignment======================================================
read dataframe

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
url_table=pd.read_html(url)
df=url_table[0]
df
```

Only process the cells that have an assigned borough

```python
df=df[df['Borough']!='Not assigned'] 
df
```

#More than one neighborhood can exist in one postal code area.More than one neighborhood can exist in one postal code area.

```python
df=df.groupby('Postal Code').agg(', '.join).reset_index()
df
#More than one neighborhood can exist in one postal code area.More than one neighborhood can exist in one postal code area.
```

If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
df['Neighborhood'][df['Neighborhood']=='Not assigned']=df['Borough'][df['Neighborhood']=='Not assigned']
df
#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
```

```python
toronto_data=dflc[dflc['Borough'].str.contains('Toronto')]
toronto_data.reset_index(drop=True, inplace=True)
latitude=toronto_data.loc[4,'Latitude']
longitude=toronto_data.loc[4,'Longitude']
#choose one at center of Toronto
```

```python
import folium
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)
for lat, lng, label in zip(dflc['Latitude'], dflc['Longitude'], dflc['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
map_toronto
# add markers to map
```
